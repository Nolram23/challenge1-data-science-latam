### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
sum1 = tienda['Precio'].sum()
sum2 = tienda2['Precio'].sum()
sum3 = tienda3['Precio'].sum()
sum4 = tienda4['Precio'].sum()
print(sum1)
print(sum2)
print(sum3)
print(sum4)
```

```python
suma_precios = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum(),
}

for tienda, suma in suma_precios.items():
    print(f"{tienda}: ${suma:,.2f}")
```

```python
cantidad_categorias = {
    'Tienda 1': tienda['Categoría del Producto'].nunique(),
    'Tienda 2': tienda2['Categoría del Producto'].nunique(),
    'Tienda 3': tienda3['Categoría del Producto'].nunique(),
    'Tienda 4': tienda4['Categoría del Producto'].nunique(),
}

for tienda, cantidad in cantidad_categorias.items():
    print(f"{tienda}: {cantidad} categorías únicas")
```

```python
# Contar cuántos productos hay por categoría
conteo_categorias = tienda['Categoría del Producto'].value_counts()
conteo_categorias2 = tienda2['Categoría del Producto'].value_counts()
conteo_categorias3 = tienda3['Categoría del Producto'].value_counts()
conteo_categorias4 = tienda4['Categoría del Producto'].value_counts()

# Unirlos en un solo DataFrame
conteo_total = pd.concat(
    [conteo_categorias, conteo_categorias2, conteo_categorias3, conteo_categorias4],
    axis=1
)

# Renombrar las columnas para identificar cada tienda
conteo_total.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Llenar valores faltantes con 0 (por si alguna tienda no tiene una categoría)
conteo_total = conteo_total.fillna(0).astype(int)

# Mostrar como tabla
print("Cantidad de ítems por categoría y tienda:\n")
print(conteo_total)
```

# 2. Ventas por categoría

```python
# Lista de las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Recorrer cada tienda y mostrar el conteo
for i, tienda in enumerate(tiendas, start=1):
    print(f"\nTienda {i}:")
    conteo = tienda['Categoría del Producto'].value_counts()
    for categoria, cantidad in conteo.items():
        print(f"- {categoria}: {cantidad} ítems")
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

# Crear un diccionario para almacenar los resultados
tabla_conteo = {}

# Recorremos las tiendas
for i, tienda in enumerate(tiendas, start=1):
    conteo = tienda['Categoría del Producto'].value_counts()
    tabla_conteo[f'Tienda {i}'] = conteo

# Convertimos el diccionario en un DataFrame, uniendo por índice (categorías)
resultado = pd.DataFrame(tabla_conteo).fillna(0).astype(int)

# Mostrar la tabla final
print(resultado)
```

# 3. Calificación promedio de la tienda


```python
# Lista de las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Crear un diccionario para almacenar los promedios
promedios = {}

# Calcular promedio de calificación por tienda
for i, tienda in enumerate(tiendas, start=1):
    promedio = tienda['Calificación'].mean()
    promedios[f'Tienda {i}'] = round(promedio, 2)  # redondeamos a 2 decimales

# Mostrar el resultado
for tienda, promedio in promedios.items():
    print(f"{tienda}: {promedio} estrellas")
```

# 4. Productos más y menos vendidos

```python
# Lista de las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Procesar cada tienda
for i, tienda in enumerate(tiendas, start=1):
    print(f"\nTienda {i}:")

    # Contar las veces que aparece cada producto
    conteo = tienda['Producto'].value_counts()

    # Producto más vendido (mayor conteo)
    producto_mas_vendido = conteo.idxmax()
    ventas_mas = conteo.max()

    # Producto menos vendido (menor conteo)
    producto_menos_vendido = conteo.idxmin()
    ventas_menos = conteo.min()

    print(f"Producto más vendido: {producto_mas_vendido} ({ventas_mas} ventas)")
    print(f"Producto menos vendido: {producto_menos_vendido} ({ventas_menos} ventas)")
```

# 5. Envío promedio por tienda

```python
# Lista de tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Procesar cada tienda y calcular el promedio
for i, tienda in enumerate(tiendas, start=1):
    promedio_envio = tienda['Costo de envío'].mean()
    print(f"Tienda {i}: Costo de envío promedio = ${promedio_envio:,.2f}")
```
